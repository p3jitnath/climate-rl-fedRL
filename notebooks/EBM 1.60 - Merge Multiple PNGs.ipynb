```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
def plot_images_in_grid(image_paths, grid_dim, output_path=None, figsize=None):
    if not figsize:
        figsize = (7, 10)
    fig, axes = plt.subplots(
        nrows=grid_dim[0], ncols=grid_dim[1], figsize=figsize
    )

    for idx, ax in enumerate(axes.flatten()):
        if idx < len(image_paths):
            img = Image.open(image_paths[idx])
            img = np.array(img)
            ax.imshow(img)
            ax.axis("off")
        else:
            ax.axis("off")

    plt.subplots_adjust(
        wspace=0.0, hspace=0.0
    )  # Adjust spacing between images
    if output_path:
        plt.savefig(output_path, dpi=300, bbox_inches="tight")
    else:
        plt.show()
```
